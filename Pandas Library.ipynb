# Handaling missing/null values

```python
import pandas as pd
```

```python
df = pd.read_csv("emp_data.csv")
df
```

```python
# check the Dtype of each columns
df.info()
```

```python
# check for null
df.isnull()
```

```python
# count the nulls for each cols
df.isnull().sum()
```

```python
# count the total nulls
df.isnull().sum()
```

```python
# check the not null values
df.notnull()
```

```python
# count notnull values in each cols
df.notnull().sum()
```

```python
df
```

```python
# drop rows that have null values
df.dropna()    #--by defualt it takes axis=0--
```

```python
# drop colums that have null values
df.dropna(axis=1)
```

```python
# return all rows that have atleast 3 non_Null values
print(df)
df.dropna(thresh= 3)
```

```python
# drop rows that have null values only in dept column
print(df)
df.dropna(subset=['Dept'])
```

```python
# drop rows that have null values only in DOJ column
print(df)
df.dropna(subset=['D.O.J'])
```

```python
# drop rows that have null values only in Dept & Salary columns
print(df)
df.dropna(subset=['Dept','Salary'])
```

```python
# insert '0' at null places in dataset
df.fillna(0)
```

```python
# insert appropriate values at null values in each columns
print(df)
df.fillna({'Dept':'Other','D.O.J':'none','Salary':'0'})
```

```python
# insert previous row values at nulls
print(df)
df.fillna(method= 'ffill')    #-- ffill = fowrard fill
```

```python
# insert next row values at nulls
print(df)
df.fillna(method= 'bfill')    #-- bfill = backward fill
```

# Pandas Groupby

```python
df=pd.read_csv('sales_data.csv')
df.head()
```

```python
# group the dataset by 'Category'
group1 = df.groupby(by = "Category")
group1.groups        #--- will return a dict having 'groupby_value' and their 'row_index' as keys
```

```python
# group the dataset by 'Category' and 'Subcategory'
df.groupby(["Category","Sub-Category"]).groups
```

```python
# Show the groups rows values of group1
for Category, df_ in group1:
    print(Category)
    print(df_)
```

```python
# Alternative 
list(group1)
```

```python
# obtain the rows that have category as 'Technology'
group2 =df.groupby('Category').get_group('Technology')
group2
```

```python
# find the total profit and total quantity for category as 'Technology'
print('Profit:',group2["Profit"].sum())
print('Quantity :',group2["Quantity"].sum())
```

```python
# find aggregate sum, mean, max for shipping cost group by Category
group1["Shipping Cost"].aggregate(['sum','mean','max'])
```

# Merging in Pandas

```python
df1 = pd.DataFrame({"ID": [101, 102, 103, 104, 105], "Name":['A','B','C','D','E']})
df2 = pd.DataFrame({"ID": [101, 102, 103, 104, 105], "Marks":[77,89,54,67,66]})
print(df1)
print(df2)
```

```python
# merge df1 and df2 where df1 
#syntax : pd.merge(left_side_dataframe, right_side_datframe, on = 'unique col')
pd.merge(df1, df2 , on = 'ID')
```
